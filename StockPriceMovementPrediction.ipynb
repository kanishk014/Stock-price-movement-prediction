#Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Reading Training Dataset

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

#Data Preprocessing

## Dropping null values data rows

```python
df = df.dropna()
```

```python
df
```

## Bringing dependent variable column at the end

```python
idcol = df['id']
idcol
```

```python
df.drop('id', inplace = True, axis = 1)
```

```python
resultcol = df['result']
resultcol
```

```python
df.drop('result', inplace = True, axis = 1)
```

```python
df.shape
```

```python
df['id'] = idcol
```

```python
df['result'] = resultcol
```

```python
df.head()
```

## Encoding all categorical columns using Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['RSI_signal'] = le.fit_transform(df['RSI_signal'])
df['Stoch_O_signal'] = le.fit_transform(df['Stoch_O_signal'])
df['ichimoku_c_signal'] = le.fit_transform(df['ichimoku_c_signal'])
df['fibonacci_signal'] = le.fit_transform(df['fibonacci_signal'])
df['Bollinger_signal'] = le.fit_transform(df['Bollinger_signal'])
```

```python
df.head(15)
```

# Splitting Data into X and y

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

# Applying One Hot Encoding to a Categorical Column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

#Splitting data into Test and Training set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 1)
```

#Applying Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Predicting the model using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0, max_depth = 15)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
